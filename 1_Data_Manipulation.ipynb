```python
import pandas as pd

# a. Read the CSV file
df = pd.read_csv('data.csv')
```

```python
# c. Show basic statistical description
description = df.describe()
```

```python
# d. Check for null values and replace with mean
df.fillna(df.mean(), inplace=True)
```

```python
# e. Select at least two columns and aggregate
selected_columns = ['Calories', 'Pulse']
agg_functions = {
    'Calories': ['min', 'max', 'count', 'mean'],
    'Pulse': ['min', 'max', 'count', 'mean']
}
aggregated_data = df[selected_columns].agg(agg_functions)
```

```python
# f. Filter rows with calories between 500 and 1000
filtered_df1 = df[(df['Calories'] >= 500) & (df['Calories'] <= 1000)]
```

```python
# g. Filter rows with calories > 500 and pulse < 100
filtered_df2 = df[(df['Calories'] > 500) & (df['Pulse'] < 100)]
```

```python
# h. Create a new dataframe without "Maxpulse" column
df_modified = df.drop(columns=['Maxpulse'])
```

```python
# i. Delete the "Maxpulse" column from the main df dataframe
df.drop(columns=['Maxpulse'], inplace=True)
```

```python
# j. Convert the datatype of Calories column to int
df['Calories'] = df['Calories'].astype(int)
```

```python
# k. Create a scatter plot for Duration and Calories
import matplotlib.pyplot as plt
df.plot.scatter(x='Duration', y='Calories')
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('Scatter Plot of Duration vs Calories')
plt.show()
```
